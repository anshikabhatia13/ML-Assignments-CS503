**Task1- K-Means Clustering**

**Concept:**


*  K-Means is a centroid-based clustering algorithm that partitions data into K clusters.
*  It iteratively assigns each data point to the nearest centroid and updates centroids until convergence.
*  The number of clusters K needs to be specified in advance.

**Steps:**



1.   Normalize the dataset (since K-Means is distance-based).
2.   Use the Elbow Method to determine the optimal number of clusters (K).
3.   Apply K-Means clustering and interpret the clusters.
* Visualize your analysis before and after clustering
4.   Visualize and analyze cluster characteristics.

**Dataset Link for K-Means Clustering**
[Mall Customer Segmentation Data](https://www.kaggle.com/datasets/vjchoudhary7/customer-segmentation-tutorial-in-python)





```python

from sklearn.metrics import pairwise_distances_argmin
def compute_wcss(X, max_k=10):
    wcss = []
    for k in range(1, max_k + 1):
        centroids = X[np.random.choice(range(X.shape[0]), k, replace=False)]
        for i in range(100):
            labels = pairwise_distances_argmin(X, centroids)
            new_centroids = np.array([X[labels == j].mean(axis=0) for j in range(k)])
            if np.all(centroids == new_centroids):
                break
            centroids = new_centroids
        wcss.append(sum(np.min((X[:, np.newaxis, :] - centroids)**2, axis=2).sum(axis=1)))
    return wcss

wcss = compute_wcss(X_scaled, max_k=10)

plt.plot(range(1, 11), wcss, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.grid(True)
plt.show()

```

```python
K=3
```

```python

def custom_kmeans(X, k, max_iters=100):
    np.random.seed(42)
    centroids = X[np.random.choice(range(X.shape[0]), k, replace=False)]

    for _ in range(max_iters):
        labels = pairwise_distances_argmin(X, centroids)
        new_centroids = np.array([X[labels == i].mean(axis=0) for i in range(k)])
        if np.allclose(centroids, new_centroids):
            break
        centroids = new_centroids
    return labels, centroids

labels, centroids = custom_kmeans(X_scaled, K)


pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap='viridis', s=40)
plt.title('Clusters After K-Means (PCA Reduced)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```

**Task-2. Hierarchical Clustering**

**Concept:**

*   Hierarchical clustering creates a tree-like structure (dendrogram) representing nested clusters.

*   It does not require predefining K.

**Two types:**

* Agglomerative (Bottom-Up): Each data point starts as its own cluster, and clusters are merged iteratively.

* Divisive (Top-Down): Starts with one large cluster, which is split recursively.

**Steps:**

1.  Normalize the dataset.

2.  Compute the distance matrix using Euclidean distance.

3.  Apply Agglomerative Clustering (most common method).

4.  Use the Dendrogram to determine the optimal number of clusters.

```python
from lxml import etree
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import normalize
from scipy.spatial.distance import pdist, squareform
import numpy as np
import os
import re
from scipy.cluster.hierarchy import dendrogram
import scipy

file_path = '/content/drive/MyDrive/ML-Lab9/dblp.xml'


entries = []
max_entries = 50
count = 0

for event, elem in etree.iterparse(file_path, events=('end',), tag=('article', 'inproceedings')):
    title = elem.findtext('title')
    year = elem.findtext('year')
    venue = elem.findtext('journal') or elem.findtext('booktitle')

    if title and year:
        entries.append({
            'title': title,
            'year': year,
            'venue': venue if venue else "Unknown"
        })
        count += 1
        if count >= max_entries:
            break
    elem.clear()

df = pd.DataFrame(entries)
print("Sample:\n", df.head())


vectorizer = TfidfVectorizer(stop_words='english', max_features=500)
X = vectorizer.fit_transform(df['title']).toarray()

X_normalized = normalize(X, norm='l2')


dist_matrix = pdist(X_normalized, metric='euclidean')


def custom_hierarchical_clustering(X):
    dist_matrix = squareform(pdist(X, metric='euclidean'))
    n = dist_matrix.shape[0]
    clusters = {i: [i] for i in range(n)}
    next_cluster_id = n
    linkage_matrix = []
    while len(clusters) > 1:
        keys = list(clusters.keys())
        min_dist = float('inf')
        to_merge = (None, None)
        for i in range(len(keys)):
            for j in range(i + 1, len(keys)):
                ci, cj = keys[i], keys[j]
                dists = [dist_matrix[p1, p2] for p1 in clusters[ci] for p2 in clusters[cj]]
                avg_dist = np.mean(dists)
                if avg_dist < min_dist:
                    min_dist = avg_dist
                    to_merge = (ci, cj)

        ci, cj = to_merge
        new_cluster = clusters[ci] + clusters[cj]
        linkage_matrix.append([ci, cj, min_dist, len(new_cluster)])
        del clusters[ci]
        del clusters[cj]
        clusters[next_cluster_id] = new_cluster
        next_cluster_id += 1

    return np.array(linkage_matrix, dtype=float)
print("Running custom hierarchical clustering...")
linkage_matrix = custom_hierarchical_clustering(X_normalized[:200])
plt.figure(figsize=(12,30 ))
dendrogram(linkage_matrix, labels=df['title'][:200].values, leaf_rotation=90, leaf_font_size=8)
plt.title(" Hierarchical Clustering Dendrogram")
plt.xlabel("Paper Titles")
plt.ylabel("Distance")
plt.tight_layout()
plt.show()


```

```python
from scipy.cluster.hierarchy import cut_tree
linkage_matrix = custom_hierarchical_clustering(X_normalized[:100])
num_clusters = 10
clusters = cut_tree(linkage_matrix, n_clusters=num_clusters).flatten()
df['cluster'] = clusters
plt.figure(figsize=(12, 6))
plt.scatter(df['title'], np.zeros_like(df['title']), c=df['cluster'], cmap='viridis')
plt.title("Cluster Visualization After Agglomerative Clustering")
plt.xlabel("Paper Titles")
plt.ylabel("Cluster")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
clustered_keywords = {}
for cluster_id in df['cluster'].unique():
    cluster_titles = df[df['cluster'] == cluster_id]['title'].values
    cluster_keywords = []
    for title in cluster_titles:
        keywords = vectorizer.transform([title]).toarray().flatten()
        cluster_keywords.append(keywords)

    clustered_keywords[cluster_id] = np.mean(cluster_keywords, axis=0)
for cluster_id, keywords in clustered_keywords.items():
    print(f"\nCluster {cluster_id} Keywords (top 5):")
    top_keywords = np.argsort(keywords)[::-1][:5]
    print([vectorizer.get_feature_names_out()[i] for i in top_keywords])
```

```python
import pandas as pd
from wordcloud import WordCloud
import re
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
import numpy as np

df = pd.read_csv('/content/training.1600000.processed.noemoticon.csv', encoding='latin-1', header=None)
df.columns = ['sentiment', 'id', 'date', 'query', 'user', 'text']

df_sample = df.sample(n=10000, random_state=42).reset_index(drop=True)
```

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

def clean_text(text):
    text = re.sub(r'http\S+', '', text)  #  URLs
    text = re.sub(r'@\w+', '', text)     #  mentions
    text = re.sub(r'#\w+', '', text)     #  hashtags
    text = re.sub(r'[^A-Za-z\s]', '', text)  #  punctuation and numbers
    text = text.lower()
    text = ' '.join([word for word in text.split() if word not in stop_words])
    return text

df_sample['clean_text'] = df_sample['text'].apply(clean_text)
```

```python
vectorizer = TfidfVectorizer(max_features=1000)
X_tfidf = vectorizer.fit_transform(df_sample['clean_text'])
```

```python
pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_tfidf.toarray())

df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2'])
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df_pca['PC1'], df_pca['PC2'], alpha=0.5)
plt.title('PCA of Tweets Before Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```

```python


neigh = NearestNeighbors(n_neighbors=5)
nbrs = neigh.fit(X_pca)
distances, indices = nbrs.kneighbors(X_pca)
distances = np.sort(distances[:, 4])
plt.figure(figsize=(8, 6))
plt.plot(distances)
plt.title('k-Distance Graph')
plt.xlabel('Data Points Sorted by Distance')
plt.ylabel('5th Nearest Neighbor Distance')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import seaborn as sns
import matplotlib.pyplot as plt

df_arm=pd.read_csv('/content/market.csv', sep=';')
df_arm.head()
```

```python
df_arm.dtypes
```

```python

scaler = StandardScaler()
data_scaled = scaler.fit_transform(df_arm)

pca = PCA(n_components=0.95)
df_arm = pca.fit_transform(data_scaled)
```

```python
df_arm.dtype
```

```python


frequent_itemsets = apriori(df_arm, min_support=0.2, use_colnames=True)
frequent_itemsets.sort_values(by='support', ascending=False).head(10)
```

```python


rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.sort_values(by='lift', ascending=False).head(10)
```

```python


top_itemsets = frequent_itemsets.nlargest(10, 'support')
sns.barplot(x='support', y=top_itemsets['itemsets'].astype(str), data=top_itemsets)
plt.title('Top 10 Frequent Itemsets by Support')
plt.xlabel('Support')
plt.ylabel('Itemsets')
plt.xticks(rotation=45)
plt.show()
```

```python
top_rules = rules.nlargest(10, 'lift')

plt.figure(figsize=(10, 6))
sns.barplot(x='lift', y=top_rules['antecedents'].astype(str) + " => " + top_rules['consequents'].astype(str), data=top_rules)
plt.title('Top 10 Association Rules by Lift')
plt.xlabel('Lift')
plt.ylabel('Rule')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
